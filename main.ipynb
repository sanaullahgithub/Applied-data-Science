## Extensive data cleaning process for the dataset
#### Steps involved:
#### 1. Removing columns with all null values.
#### 2. Filling missing values where appropriate.
#### 3. Removing duplicate rows.
#### 4. Correcting data types.
#### 5. Handling outliers or anomalies if any.

```python


# Re-reading the original dataset as a precaution
data_api = pd.read_csv(data_file, skiprows=4)

# Step 1: Remove columns with all null values
data_api = data_api.dropna(axis=1, how='all')

# Step 2: Fill missing values
# For numerical columns (years), replace NaN with 0 (assuming no data means 0% renewable energy consumption)
year_columns = data_api.columns[4:]  # Assuming year data starts from the 5th column
data_api[year_columns] = data_api[year_columns].fillna(0)

# Step 3: Remove duplicate rows
data_api = data_api.drop_duplicates()

# Step 4: Correct data types
# Converting year columns to float (as they represent percentages)
data_api[year_columns] = data_api[year_columns].astype(float)

# Step 5: Handling outliers or anomalies
for year in year_columns:
    mean_value = data_api[year].mean()
    std_dev = data_api[year].std()
    outlier_threshold = mean_value + 3 * std_dev
    data_api[year] = data_api[year].apply(lambda x: min(x, outlier_threshold))

# Display the cleaned data
data_api.head()
```

```python
# Data Cleaning

# Dropping unnecessary columns
cleaned_data_df = data_api.drop(['Indicator Name', 'Indicator Code', 'Unnamed: 67'], axis=1)

# Handling missing values: Since it's a time series data, we can fill missing values using interpolation
cleaned_data_df.iloc[:, 3:] = cleaned_data_df.iloc[:, 3:].interpolate(axis=1)

# Checking the cleaned data
cleaned_data_df.head()
```

## Model training and predictions

```python
# Selecting one country from each cluster for the analysis
# We will select the first country in each cluster for simplicity
selected_countries = cluster_data.reset_index().groupby('Cluster')['Country Name'].first()
selected_countries_list = selected_countries.tolist()

# Preparing data for curve fitting - focusing on the years 2000-2020
fit_years = [str(year) for year in range(2000, 2021)]
fit_data = cleaned_data_df[cleaned_data_df['Country Name'].isin(selected_countries_list)]
fit_data = fit_data[['Country Name'] + fit_years].set_index('Country Name')

fit_data.head()
```

```python
from scipy.optimize import curve_fit

# Define the models for curve fitting

# Exponential growth model
def exp_growth(x, a, b, c):
    return a * np.exp(b * x) + c

# Polynomial model (second order)
def poly_2nd_order(x, a, b, c):
    return a * x**2 + b * x + c

# Function to fit model and predict future values
def fit_and_predict(data, model_func, years_to_predict):
    x_data = np.array(range(len(data)))
    y_data = data.values

    # Fit the model
    params, _ = curve_fit(model_func, x_data, y_data, maxfev=10000)

    # Predict future values
    future_x = np.array(range(len(data) + years_to_predict))
    future_y = model_func(future_x, *params)

    return future_x, future_y, params

# Years to predict (10 and 20 years into the future)
years_to_predict = 20

# Fitting models and predicting for each selected country
predictions = {}
for country in selected_countries_list:
    data = fit_data.loc[country]
    future_x, future_y, _ = fit_and_predict(data, exp_growth, years_to_predict)
    predictions[country] = (future_x, future_y)

# Displaying the first set of predictions
predictions[selected_countries_list[0]]
```

```python
import numpy as np

# Defining the err_ranges function to estimate confidence ranges
def err_ranges(params, cov, x):
    """ Calculate the error ranges for the fitted function at given x values. """
    # Partial derivatives of the function with respect to the parameters
    partials = np.array([x**i for i in range(len(params))])
    # Propagation of error
    sigma_y = np.sqrt(np.sum((cov @ partials)**2, axis=0))
    return sigma_y

# Visualizing the best fitting function and the confidence range for each country
plt.figure(figsize=(15, 10))

for i, country in enumerate(selected_countries_list):
    data = fit_data.loc[country]
    x_data = np.array(range(len(data)))
    y_data = data.values

    # Fit the model
    params, cov = curve_fit(exp_growth, x_data, y_data, maxfev=10000)

    # Predict future values
    future_x = np.array(range(len(data) + years_to_predict))
    future_y = exp_growth(future_x, *params)

    # Error ranges
    sigma_y = err_ranges(params, cov, future_x)

    # Plotting
    plt.subplot(3, 2, i+1)
    plt.plot(future_x[:len(data)], y_data, 'o', label=f'Actual Data ({country})')
    plt.plot(future_x, future_y, '-', label='Fitted Model')
    plt.fill_between(future_x, future_y - sigma_y, future_y + sigma_y, alpha=0.2)
    plt.title(country)
    plt.xlabel('Years since 2000')
    plt.ylabel('Renewable Energy Consumption (%)')
    plt.legend()

plt.tight_layout()
plt.show()
```
